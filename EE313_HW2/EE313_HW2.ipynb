```python
#-------------------------------------------------------------------------------
# Sophia Xu
# shx59
# EE313 - Hall
# HW2
# 2/6/2020
#--------------------------------------------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv("SP500.csv")

data['Day'] = data.index
#print(data.head())
data_list = list(data['SP500'])

avg_3 = []
for n in range(2, len(data.index)):
    calc = (1/3)*(data.at[n, 'SP500'] + data.at[n-1, 'SP500'] + data.at[n-2, 'SP500'])
    avg_3.append(calc)
run_avg_3 = pd.DataFrame(avg_3, columns = ['3 Day Moving Average'])
run_avg_3.index += 2


avg_15 = []
for n in range(14, len(data.index)):
    summing = 0
    for k in range(0, 15):
        summing += data.at[n-k, 'SP500']
        #print(summing)
    summing = summing * (1/15)
    avg_15.append(summing)

run_avg_15 = pd.DataFrame(avg_15, columns = ['15 Day Moving Average'])
run_avg_15.index += 14


result = pd.concat([data, run_avg_3, run_avg_15], axis=1, sort=False)
hello = result.plot(kind = 'line', x = 'Day', title = 'S&P 500 Closing Price Data, Past 503 Work Days')
hello.set_xlabel("Work Day")
hello.set_ylabel("Stock Price Data($)")
plt.show()


# def running_avg_fun(n, data):
#     # ex: n = 15
#     index_to_subtract = n-1
#     average = []
#     for n1 in range(n, len(data.index)):
#         # calculate the summation
#         summing = 0
#         for k in range(0, index_to_subtract+1):
#             summing += data.at[k, 'SP500']
            
#         summing = summing * (1/n)
#         average.append(summing)
    
#     avg_run = pd.DataFrame(average, columnds = ['Moving Average'])
#     avg_run.index += n
#     avg_run['Day'] = avg_run.index
#     return avg_run
```
