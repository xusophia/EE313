```python
import matplotlib.pyplot as plt
import math
import numpy as np
t = np.linspace(0,1,1000)
x = np.cos(2*math.pi*t)
# Create the plot
plt.plot(t,x)
plt.show()


```

```python
y = np.real(np.e ** (1j*2*math.pi*t))
plt.plot(t,y)
plt.show()
```

```python
z = np.real(((np.e ** (1j*2*math.pi*t)) + (np.e ** (-1j*2*math.pi*t)))/2)
plt.plot(t, z)
plt.show()
```
